```python
#Dependencies
from config import tkey
from config import tskey
import pandas as pd
import tweepy
import csv
```

```python
auth = tweepy.OAuthHandler(tkey, tskey)
```

```python
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
#Find tweets and write them to csv
searchterm = "#brooklyn"
brooklyncsv = open('t_brooklyn.csv', 'a')
csvWriter = csv.writer(brooklyncsv)

for tweet in tweepy.Cursor(api.search, q=searchterm + " -filter:retweets", count=100, lang="en", 
                           since="2019-10-01", tweet_mode='extended').items():
    print (tweet.created_at, tweet.full_text)
    csvWriter.writerow([tweet.created_at, tweet.full_text.encode('utf-8')])
```
