```python
from config import tkey
from config import tskey
import pandas as pd
import tweepy
import csv
```

```python
auth = tweepy.OAuthHandler(tkey, tskey)
```

```python
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
brooklyncsv = open('brooklyn.csv', 'a')

csvWriter = csv.writer(brooklyncsv)

for tweet in tweepy.Cursor(api.search,q="#brooklyn",count=100,
                           lang="en",
                           since="2019-01-01").items():
    print (tweet.created_at, tweet.text)
    csvWriter.writerow([tweet.created_at, tweet.text.encode('utf-8')])
```
