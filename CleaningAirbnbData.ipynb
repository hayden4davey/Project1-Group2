```python
#Dependencies
import pandas as pd

#Read data and convert to data frame
abb_data = pd.read_csv("AB_NYC_2019.csv")
abb_df = pd.DataFrame(abb_data)
abb_df.head()
```

```python
#Check for duplicates
abb_df.duplicated().sum()
```

```python
#Check for Airbnb's availability for 2019
abb_df['availability_365'].value_counts().head()
```

```python
#Drop Airbnb's with no availability for 2019
unavail_abb_df = abb_df[abb_df['availability_365']==0]
clean_abb_df = abb_df[abb_df['availability_365']>0]

len(clean_abb_df)
```

```python
#Check for nulls
clean_abb_df.isnull().sum()
```

```python
#Check data types
clean_abb_df.dtypes
```

```python
#Check data frame
clean_abb_df.head()
```

```python
#Rename columns
column_dict = {'id':'ID', 'name':'Name', 'host_id':'Host ID', 'host_name':'Host Name','neighbourhood':'Neighbourhood',
'latitude':'Latitude','longtiude':'Longitude','room_type':'Room Type','price':'Price',
'minimum_nights':'Minimum Nights', 'number_of_reviews':'Number of Reviews', 'last_review':'Last Review',
'reviews_per_month':'Reviews per Month','calculated_host_listings_count':'Calculated Host Listings Count',
'availability_365':'Availability'}

clean_abb_df = clean_abb_df.rename(column_dict, axis=1).head()
```

```python
#Export to CSV
clean_abb_df.to_csv("CleanAirbnbData.csv", index=False, header=True)
clean_abb_df.head()
```
