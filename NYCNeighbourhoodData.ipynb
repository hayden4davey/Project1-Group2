```python
%matplotlib inline 


#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Read neighbourhood data and convert to data frame
nbhd_data = pd.read_csv("Resources/New_York_City_Population_By_Neighborhood_Tabulation_Areas.csv")
nbhd_df = pd.DataFrame(nbhd_data[nbhd_data['Year']==2010])
nbhd_df = nbhd_df.rename(columns={'NTA Name':'Neighbourhood'})
nbhd_df.head()
```

```python
#Read clean Airbnb data and convert to data frame
clean_abb_data = pd.read_csv("Resources/CleanAirbnbData.csv")
clean_abb_df = pd.DataFrame(clean_abb_data)
clean_abb_df
```

```python
#Count number of listings and make it into a df
nbhd_listings = pd.DataFrame(clean_abb_df['Neighbourhood'].value_counts())

nbhd_listings = nbhd_listings.reset_index()
nbhd_listings = nbhd_listings.rename(columns={'index':'Neighbourhood','Neighbourhood':'Listings Count'})
nbhd_listings.head()
```

```python
#Merge dfs
combined_df = nbhd_listings.merge(nbhd_df, how='inner', on='Neighbourhood')

combined_df
```

```python
combined_df.plot.scatter(x='Population', y='Listings Count')
```

```python
sns.regplot(data=combined_df, x='Population', y='Listings Count')
```
